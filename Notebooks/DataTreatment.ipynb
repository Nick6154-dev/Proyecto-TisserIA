```python
import os
from PIL import Image

# Definición de las carpetas de origen y destino
principal_folder = 'C:\\Workspace\\Dispositivos Moviles\\Proyecto\\Notebooks\\clothes_and_no-clothes'
destination_folder = 'C:\\Workspace\\Dispositivos Moviles\\Proyecto\\Notebooks\\dataset'

# Obtener una lista de carpetas dentro de la carpeta principal
list_folder = os.listdir(principal_folder)

# Iterar sobre cada carpeta dentro de la carpeta principal
for folder in list_folder:
    # Construir la ruta completa de la carpeta actual
    folder_path = os.path.join(principal_folder, folder)

    # Obtener una lista de archivos dentro de la carpeta actual
    list_sub_folder = os.listdir(folder_path)

    # Crear la carpeta de destino correspondiente si no existe
    destination_sub_folder = os.path.join(destination_folder, folder)
    if not os.path.exists(destination_sub_folder):
        os.makedirs(destination_sub_folder)

    # Inicializar un contador para asegurar que los nombres sean únicos dentro de cada carpeta
    counter = 1

    # Iterar sobre cada archivo dentro de la carpeta actual, ordenados alfabéticamente
    for index, image in enumerate(sorted(list_sub_folder)):
        # Construir la ruta completa del archivo actual
        image_path = os.path.join(folder_path, image)

        # Verificar si el archivo no es un archivo PDF
        if not image.endswith('.pdf'):
            # Abrir la imagen con Pillow
            with Image.open(image_path) as img:
                # Redimensionar la imagen a 255x255
                img = img.resize((255, 255))

                # Convertir la imagen al espacio de color RGB
                img = img.convert('RGB')

                # Establecer el nuevo nombre base para el archivo, usando el nombre de la carpeta actual
                new_image_name = f"{folder}_{counter}.jpg"

                # Construir la ruta completa de destino para el archivo
                destination_path = os.path.join(destination_sub_folder, new_image_name)

                # Guardar la imagen redimensionada y convertida en RGB
                img.save(destination_path)

                # Aumentar el contador para aumentar la secuencia
                counter += 1
```

```python
import os
import fitz  # PyMuPDF

principal_folder = './images-pdf'
destination_folder = './crochet-images'

# Lista de carpetas en el directorio principal
list_folder = os.listdir(principal_folder)

# Iterar sobre cada carpeta en el directorio principal
for folder in list_folder:
    folder_path = os.path.join(principal_folder, folder)
    list_sub_folder = os.listdir(folder_path)

    # Inicializar contador para las imágenes
    image_counter = 1

    # Iterar sobre cada imagen en la carpeta
    for image in list_sub_folder:
        image_path = os.path.join(folder_path, image)

        # Verificar si el archivo es una imagen
        if not image.endswith('.pdf'):
            # Construir nuevo nombre de la imagen
            new_image_name = f"{image_counter}.jpg"

            # Ruta de destino para la imagen renombrada
            destination_path = os.path.join(destination_folder, new_image_name)

            # Copiar la imagen y cambiarle el nombre
            os.system(f"cp \"{image_path}\" \"{destination_path}\"")

            # Incrementar contador de imágenes
            image_counter += 1

        # Si el archivo es un PDF, extraer imágenes y etiquetarlas
        elif image.endswith('.pdf'):
            # Abre el archivo PDF
            pdf_path = os.path.join(folder_path, image)
            print(pdf_path)
            pdf_document = fitz.open(pdf_path)

            # Iterar sobre cada página del PDF
            for page_number in range(len(pdf_document)):
                # Extraer imágenes de la página
                page = pdf_document.load_page(page_number)
                images = page.get_images(full=True)

                # Iterar sobre cada imagen de la página
                for image_index, img in enumerate(images):
                    xref = img[0]
                    base_image = pdf_document.extract_image(xref)
                    image_bytes = base_image["image"]
                    image_extension = base_image["ext"]
                    image_name = f"{image_counter}_{image_index+1}.{image_extension}"

                    # Guardar la imagen extraída con el nuevo nombre
                    image_path = os.path.join(destination_folder, image_name)
                    with open(image_path, "wb") as f:
                        f.write(image_bytes)

            # Cerrar el documento PDF
            pdf_document.close()
```

```python
import os
import fitz  # PyMuPDF
import json
import shutil
from PIL import Image
import re

principal_folder = 'C:\Workspace\Dispositivos Moviles\Proyecto\Dataset\images-pdf'
train_folder = './train'
metadata_file = os.path.join(train_folder, 'metadata.jsonl')
destination_folder = './train'
count = 1

# Lista de carpetas en el directorio principal
list_folder = os.listdir(principal_folder)

# Crear la carpeta 'train' si no existe
if not os.path.exists(train_folder):
    os.makedirs(train_folder)

# Crear la carpeta 'train' si no existe
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# Iterar sobre cada carpeta en el directorio principal
for folder in list_folder:
    folder_path = os.path.join(principal_folder, folder)
    list_sub_folder = os.listdir(folder_path)

    # Iterar sobre cada archivo en la carpeta
    for file_name in list_sub_folder:
        file_path = os.path.join(folder_path, file_name)
        destination_path = os.path.join(destination_folder, file_name)

        # Construir nuevo nombre de la imagen
        new_image_name = f"crochet.{count}.jpg"

        # Verificar si el archivo es una imagen
        if file_name.endswith('.pdf'):
            try:
                # Intentar abrir el archivo PDF y extraer su contenido de texto
                pdf_document = fitz.open(file_path)
                text_content = ""
                for page in pdf_document:
                    text_content += page.get_text()
                # Cerrar el documento PDF
                pdf_document.close()

                # Eliminar saltos de línea, espacios y otros caracteres no deseados usando expresiones regulares
                text_content = re.sub(r'\s+', ' ', text_content)

                # Guardar el contenido del PDF en el archivo JSONL
                with open(metadata_file, 'a') as jsonl_file:
                    json_data = {
                        "file_name": new_image_name,
                        "additional_feature": text_content
                    }
                    jsonl_file.write(json.dumps(json_data) + '\n')

                # Guardar la imagen correspondiente
                jpg_file_path = os.path.join(folder_path, file_name.replace('.pdf', '.jpg'))
                new_image_path = os.path.join(destination_folder, new_image_name)
                if os.path.exists(jpg_file_path):
                    img = Image.open(jpg_file_path)
                    img = img.convert('RGB')
                    img = img.resize((224,224))
                    img.save(new_image_path)
                count += 1

            except Exception as e:
                print(f"Archivo danado?: {str(e)}")
```
